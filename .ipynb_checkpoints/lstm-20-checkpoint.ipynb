```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense, Input
from keras.optimizers import Adam
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
```

```python
data = pd.read_csv('dataset.csv')
```

```python
data.head()
```

```python
def create_dataset(X, y, time_steps=1):
    Xs, ys = [], []
    for i in range(len(X) - time_steps + 1):
        Xs.append(X[i:(i + time_steps)])
        ys.append(y[i + time_steps - 1])
    return np.array(Xs), np.array(ys)
```

```python
results = []
mae_results = []
mse_results = []
mape_results = []
```

```python
# for name in unique_names:
#     province_data = data[data['name'] == name]

#     X = province_data[['tanggal', 'bulan', 'tahun']].values
#     y_temp = province_data['temp_celsius'].values
#     y_humidity = province_data['humidity'].values
#     y_cloudcover = province_data['cloudcover'].values
#     y_sealevelpressure = province_data['sealevelpressure'].values
#     y_windspeed = province_data['windspeed'].values
#     y_precip = province_data['precip'].values

#     X_train, X_test, y_temp_train, y_temp_test = train_test_split(X, y_temp, test_size=0.2, random_state=42)
#     _, _, y_humidity_train, y_humidity_test = train_test_split(X, y_humidity, test_size=0.2, random_state=42)
#     _, _, y_cloudcover_train, y_cloudcover_test = train_test_split(X, y_cloudcover, test_size=0.2, random_state=42)
#     _, _, y_sealevelpressure_train, y_sealevelpressure_test = train_test_split(X, y_sealevelpressure, test_size=0.2, random_state=42)
#     _, _, y_windspeed_train, y_windspeed_test = train_test_split(X, y_windspeed, test_size=0.2, random_state=42)
#     _, _, y_precip_train, y_precip_test = train_test_split(X, y_precip, test_size=0.2, random_state=42)


#     time_steps = 1 
#     X_temp_train, y_temp_train = create_dataset(X_train, y_temp_train, time_steps)
#     X_humidity_train, y_humidity_train = create_dataset(X_train, y_humidity_train, time_steps)
#     X_cloudcover_train, y_cloudcover_train = create_dataset(X_train, y_cloudcover_train, time_steps)
#     X_sealevelpressure_train, y_sealevelpressure_train = create_dataset(X_train, y_sealevelpressure_train, time_steps)
#     X_windspeed_train, y_windspeed_train = create_dataset(X_train, y_windspeed_train, time_steps)
#     X_precip_train, y_precip_train = create_dataset(X_train, y_precip_train, time_steps)

#     X_temp_train = X_temp_train.reshape((X_temp_train.shape[0], X_temp_train.shape[1], 3))  
#     X_humidity_train = X_humidity_train.reshape((X_humidity_train.shape[0], X_humidity_train.shape[1], 3))
#     X_cloudcover_train = X_cloudcover_train.reshape((X_cloudcover_train.shape[0], X_cloudcover_train.shape[1], 3))
#     X_sealevelpressure_train = X_sealevelpressure_train.reshape((X_sealevelpressure_train.shape[0], X_sealevelpressure_train.shape[1], 3))
#     X_windspeed_train = X_windspeed_train.reshape((X_windspeed_train.shape[0], X_windspeed_train.shape[1], 3))
#     X_precip_train = X_precip_train.reshape((X_precip_train.shape[0], X_precip_train.shape[1], 3))

#     def build_model():
#         model = Sequential()
#         model.add(Input(shape=(time_steps, 3)))  # Menggunakan Input sebagai lapisan pertama
#         model.add(LSTM(50, activation='relu'))
#         model.add(Dense(1))
#         model.compile(optimizer=Adam(), loss='mse')
#         return model

#     lstm_temp = build_model()
#     lstm_humidity = build_model()
#     lstm_cloudcover = build_model()
#     lstm_sealevelpressure = build_model()
#     lstm_windspeed = build_model()
#     lstm_precip = build_model()

#     lstm_temp.fit(X_temp_train, y_temp_train, epochs=10, verbose=0)
#     lstm_humidity.fit(X_humidity_train, y_humidity_train, epochs=10, verbose=0)
#     lstm_cloudcover.fit(X_cloudcover_train, y_cloudcover_train, epochs=10, verbose=0)
#     lstm_sealevelpressure.fit(X_sealevelpressure_train, y_sealevelpressure_train, epochs=10, verbose=0)
#     lstm_windspeed.fit(X_windspeed_train, y_windspeed_train, epochs=10, verbose=0)
#     lstm_precip.fit(X_precip_train, y_precip_train, epochs=10, verbose=0)

#     X_temp_test, _ = create_dataset(X_test, y_temp_test, time_steps)
#     X_humidity_test, _ = create_dataset(X_test, y_humidity_test, time_steps)
#     X_cloudcover_test, _ = create_dataset(X_test, y_cloudcover_test, time_steps)
#     X_sealevelpressure_test, _ = create_dataset(X_test, y_sealevelpressure_test, time_steps)
#     X_windspeed_test, _ = create_dataset(X_test, y_windspeed_test, time_steps)
#     X_precip_test, _ = create_dataset(X_test, y_precip_test, time_steps)

#     X_temp_test = X_temp_test.reshape((X_temp_test.shape[0], X_temp_test.shape[1], 3))
#     X_humidity_test = X_humidity_test.reshape((X_humidity_test.shape[0], X_humidity_test.shape[1], 3))
#     X_cloudcover_test = X_cloudcover_test.reshape((X_cloudcover_test.shape[0], X_cloudcover_test.shape[1], 3))
#     X_sealevelpressure_test = X_sealevelpressure_test.reshape((X_sealevelpressure_test.shape[0], X_sealevelpressure_test.shape[1], 3))
#     X_windspeed_test = X_windspeed_test.reshape((X_windspeed_test.shape[0], X_windspeed_test.shape[1], 3))
#     X_precip_test = X_precip_test.reshape((X_precip_test.shape[0], X_precip_test.shape[1], 3))

#     temp_pred_test = lstm_temp.predict(X_temp_test)
#     humidity_pred_test = lstm_humidity.predict(X_humidity_test)
#     cloudcover_pred_test = lstm_cloudcover.predict(X_cloudcover_test)
#     sealevelpressure_pred_test = lstm_sealevelpressure.predict(X_sealevelpressure_test)
#     windspeed_pred_test = lstm_windspeed.predict(X_windspeed_test)
#     precip_pred_test = lstm_precip.predict(X_precip_test)

#     # Menghitung MAE
#     mae_temp = mean_absolute_error(y_temp_test, temp_pred_test)
#     mae_humidity = mean_absolute_error(y_humidity_test, humidity_pred_test)
#     mae_cloudcover = mean_absolute_error(y_cloudcover_test, cloudcover_pred_test)
#     mae_sealevelpressure = mean_absolute_error(y_sealevelpressure_test, sealevelpressure_pred_test)
#     mae_windspeed = mean_absolute_error(y_windspeed_test, windspeed_pred_test)
#     mae_precip = mean_absolute_error(y_precip_test, precip_pred_test)

#     # Menghitung MSE
#     mse_temp = mean_squared_error(y_temp_test, temp_pred_test)
#     mse_humidity = mean_squared_error(y_humidity_test, humidity_pred_test)
#     mse_cloudcover = mean_squared_error(y_cloudcover_test, cloudcover_pred_test)
#     mse_sealevelpressure = mean_squared_error(y_sealevelpressure_test, sealevelpressure_pred_test)
#     mse_windspeed = mean_squared_error(y_windspeed_test, windspeed_pred_test)
#     mse_precip = mean_squared_error(y_precip_test, precip_pred_test)

#     # Menghitung MAPE
#     mape_temp = mean_absolute_percentage_error(y_temp_test, temp_pred_test)
#     mape_humidity = mean_absolute_percentage_error(y_humidity_test, humidity_pred_test)
#     mape_cloudcover = mean_absolute_percentage_error(y_cloudcover_test, cloudcover_pred_test)
#     mape_sealevelpressure = mean_absolute_percentage_error(y_sealevelpressure_test, sealevelpressure_pred_test)
#     mape_windspeed = mean_absolute_percentage_error(y_windspeed_test, windspeed_pred_test)
#     mape_precip = mean_absolute_percentage_error(y_precip_test, precip_pred_test)

#     mae_results.append({
#         'name': name,
#         'mae_temp': mae_temp.round(3),
#         'mae_humidity': mae_humidity.round(3),
#         'mae_cloudcover': mae_cloudcover.round(3),
#         'mae_sealevelpressure': mae_sealevelpressure.round(3),
#         'mae_windspeed': mae_windspeed.round(3),
#         'mae_precip': mae_precip.round(3)
#     })

#     mse_results.append({
#         'name': name,
#         'mse_temp': mse_temp.round(3),
#         'mse_humidity': mse_humidity.round(3),
#         'mse_cloudcover': mse_cloudcover.round(3),
#         'mse_sealevelpressure': mse_sealevelpressure.round(3),
#         'mse_windspeed': mse_windspeed.round(3),
#         'mse_precip': mse_precip.round(3)
#     })

#     mape_results.append({
#         'name': name,
#         'mape_temp': mape_temp.round(3),
#         'mape_humidity': mape_humidity.round(3),
#         'mape_cloudcover': mape_cloudcover.round(3),
#         'mape_sealevelpressure': mape_sealevelpressure.round(3),
#         'mape_windspeed': mape_windspeed.round(3),
#         'mape_precip': mape_precip.round(3)
#     })

#     date_range = pd.date_range(start='2024-01-01', end='2030-12-30')
#     future_dates = pd.DataFrame({
#         'tanggal': date_range.day,
#         'bulan': date_range.month,
#         'tahun': date_range.year,
#         'name': name
#     })

#     future_X = future_dates[['tanggal', 'bulan', 'tahun']]
#     future_X = future_X.to_numpy().reshape((future_X.shape[0], time_steps, 3))

#     future_temp = lstm_temp.predict(future_X)
#     future_humidity = lstm_humidity.predict(future_X)
#     future_cloudcover = lstm_cloudcover.predict(future_X)
#     future_sealevelpressure = lstm_sealevelpressure.predict(future_X)
#     future_windspeed = lstm_windspeed.predict(future_X)
#     future_precip = lstm_precip.predict(future_X)

#     results.append({
#         'name': name,
#         'future_dates': date_range,
#         'future_temp': future_temp.flatten(),
#         'future_humidity': future_humidity.flatten(),
#         'future_cloudcover': future_cloudcover.flatten(),
#         'future_sealevelpressure': future_sealevelpressure.flatten(),
#         'future_windspeed': future_windspeed.flatten(),
#         'future_precip': future_precip.flatten()
#     })

for name in unique_names:
    province_data = data[data['name'] == name]

    # Inisialisasi scaler untuk X dan setiap variabel y
    scaler_X = MinMaxScaler()
    scaler_temp = MinMaxScaler()
    scaler_humidity = MinMaxScaler()
    scaler_cloudcover = MinMaxScaler()
    scaler_sealevelpressure = MinMaxScaler()
    scaler_windspeed = MinMaxScaler()
    scaler_precip = MinMaxScaler()

    # Normalisasi data input X (tanggal, bulan, tahun)
    X = scaler_X.fit_transform(province_data[['tanggal', 'bulan', 'tahun']].values)


    # Normalisasi setiap variabel y
    y_temp = scaler_temp.fit_transform(province_data['temp_celsius'].values.reshape(-1, 1)).flatten()
    y_humidity = scaler_humidity.fit_transform(province_data['humidity'].values.reshape(-1, 1)).flatten()
    y_cloudcover = scaler_cloudcover.fit_transform(province_data['cloudcover'].values.reshape(-1, 1)).flatten()
    y_sealevelpressure = scaler_sealevelpressure.fit_transform(province_data['sealevelpressure'].values.reshape(-1, 1)).flatten()
    y_windspeed = scaler_windspeed.fit_transform(province_data['windspeed'].values.reshape(-1, 1)).flatten()
    y_precip = scaler_precip.fit_transform(province_data['precip'].values.reshape(-1, 1)).flatten()

    # Split data menjadi train dan test
    X_train, X_test, y_temp_train, y_temp_test = train_test_split(X, y_temp, test_size=0.2, random_state=42)
    _, _, y_humidity_train, y_humidity_test = train_test_split(X, y_humidity, test_size=0.2, random_state=42)
    _, _, y_cloudcover_train, y_cloudcover_test = train_test_split(X, y_cloudcover, test_size=0.2, random_state=42)
    _, _, y_sealevelpressure_train, y_sealevelpressure_test = train_test_split(X, y_sealevelpressure, test_size=0.2, random_state=42)
    _, _, y_windspeed_train, y_windspeed_test = train_test_split(X, y_windspeed, test_size=0.2, random_state=42)
    _, _, y_precip_train, y_precip_test = train_test_split(X, y_precip, test_size=0.2, random_state=42)

    # Buat dataset time series untuk LSTM
    time_steps = 1
    X_temp_train, y_temp_train = create_dataset(X_train, y_temp_train, time_steps)
    X_humidity_train, y_humidity_train = create_dataset(X_train, y_humidity_train, time_steps)
    X_cloudcover_train, y_cloudcover_train = create_dataset(X_train, y_cloudcover_train, time_steps)
    X_sealevelpressure_train, y_sealevelpressure_train = create_dataset(X_train, y_sealevelpressure_train, time_steps)
    X_windspeed_train, y_windspeed_train = create_dataset(X_train, y_windspeed_train, time_steps)
    X_precip_train, y_precip_train = create_dataset(X_train, y_precip_train, time_steps)

    # Reshape data untuk input LSTM
    X_temp_train = X_temp_train.reshape((X_temp_train.shape[0], X_temp_train.shape[1], 3))
    X_humidity_train = X_humidity_train.reshape((X_humidity_train.shape[0], X_humidity_train.shape[1], 3))
    X_cloudcover_train = X_cloudcover_train.reshape((X_cloudcover_train.shape[0], X_cloudcover_train.shape[1], 3))
    X_sealevelpressure_train = X_sealevelpressure_train.reshape((X_sealevelpressure_train.shape[0], X_sealevelpressure_train.shape[1], 3))
    X_windspeed_train = X_windspeed_train.reshape((X_windspeed_train.shape[0], X_windspeed_train.shape[1], 3))
    X_precip_train = X_precip_train.reshape((X_precip_train.shape[0], X_precip_train.shape[1], 3))

    # Membangun model
    def build_model():
        model = Sequential()
        model.add(Input(shape=(time_steps, 3)))
        model.add(LSTM(50, activation='relu'))
        model.add(Dense(1))
        model.compile(optimizer=Adam(), loss='mse')
        return model

    # Inisialisasi model untuk masing-masing variabel target
    lstm_temp = build_model()
    lstm_humidity = build_model()
    lstm_cloudcover = build_model()
    lstm_sealevelpressure = build_model()
    lstm_windspeed = build_model()
    lstm_precip = build_model()

    # Training setiap model
    lstm_temp.fit(X_temp_train, y_temp_train, epochs=20, verbose=0)
    lstm_humidity.fit(X_humidity_train, y_humidity_train, epochs=20, verbose=0)
    lstm_cloudcover.fit(X_cloudcover_train, y_cloudcover_train, epochs=20, verbose=0)
    lstm_sealevelpressure.fit(X_sealevelpressure_train, y_sealevelpressure_train, epochs=20, verbose=0)
    lstm_windspeed.fit(X_windspeed_train, y_windspeed_train, epochs=20, verbose=0)
    lstm_precip.fit(X_precip_train, y_precip_train, epochs=20, verbose=0)

    # Test dataset
    X_temp_test, y_temp_test = create_dataset(X_test, y_temp_test, time_steps)
    X_humidity_test, y_humidity_test = create_dataset(X_test, y_humidity_test, time_steps)
    X_cloudcover_test, y_cloudcover_test = create_dataset(X_test, y_cloudcover_test, time_steps)
    X_sealevelpressure_test, y_sealevelpressure_test = create_dataset(X_test, y_sealevelpressure_test, time_steps)
    X_windspeed_test, y_windspeed_test = create_dataset(X_test, y_windspeed_test, time_steps)
    X_precip_test, y_precip_test = create_dataset(X_test, y_precip_test, time_steps)

    X_temp_test = X_temp_test.reshape((X_temp_test.shape[0], X_temp_test.shape[1], 3))
    X_humidity_test = X_humidity_test.reshape((X_humidity_test.shape[0], X_humidity_test.shape[1], 3))
    X_cloudcover_test = X_cloudcover_test.reshape((X_cloudcover_test.shape[0], X_cloudcover_test.shape[1], 3))
    X_sealevelpressure_test = X_sealevelpressure_test.reshape((X_sealevelpressure_test.shape[0], X_sealevelpressure_test.shape[1], 3))
    X_windspeed_test = X_windspeed_test.reshape((X_windspeed_test.shape[0], X_windspeed_test.shape[1], 3))
    X_precip_test = X_precip_test.reshape((X_precip_test.shape[0], X_precip_test.shape[1], 3))

    # Prediksi dan denormalisasi hasil
    temp_pred_test = scaler_temp.inverse_transform(lstm_temp.predict(X_temp_test))
    humidity_pred_test = scaler_humidity.inverse_transform(lstm_humidity.predict(X_humidity_test))
    cloudcover_pred_test = scaler_cloudcover.inverse_transform(lstm_cloudcover.predict(X_cloudcover_test))
    sealevelpressure_pred_test = scaler_sealevelpressure.inverse_transform(lstm_sealevelpressure.predict(X_sealevelpressure_test))
    windspeed_pred_test = scaler_windspeed.inverse_transform(lstm_windspeed.predict(X_windspeed_test))
    precip_pred_test = scaler_precip.inverse_transform(lstm_precip.predict(X_precip_test))

    y_temp_test = scaler_temp.inverse_transform(y_temp_test.reshape(-1, 1))
    y_humidity_test = scaler_humidity.inverse_transform(y_humidity_test.reshape(-1, 1))
    y_cloudcover_test = scaler_cloudcover.inverse_transform(y_cloudcover_test.reshape(-1, 1))
    y_sealevelpressure_test = scaler_sealevelpressure.inverse_transform(y_sealevelpressure_test.reshape(-1, 1))
    y_windspeed_test = scaler_windspeed.inverse_transform(y_windspeed_test.reshape(-1, 1))
    y_precip_test = scaler_precip.inverse_transform(y_precip_test.reshape(-1, 1))

    # Hitung metrik error
    mae_results.append({
        'name': name,
        'mae_temp': mean_absolute_error(y_temp_test, temp_pred_test).round(3),
        'mae_humidity': mean_absolute_error(y_humidity_test, humidity_pred_test).round(3),
        'mae_cloudcover': mean_absolute_error(y_cloudcover_test, cloudcover_pred_test).round(3),
        'mae_sealevelpressure': mean_absolute_error(y_sealevelpressure_test, sealevelpressure_pred_test).round(3),
        'mae_windspeed': mean_absolute_error(y_windspeed_test, windspeed_pred_test).round(3),
        'mae_precip': mean_absolute_error(y_precip_test, precip_pred_test).round(3),
    })

    mse_results.append({
        'name': name,
        'mse_temp': mean_squared_error(y_temp_test, temp_pred_test).round(3),
        'mse_humidity': mean_squared_error(y_humidity_test, humidity_pred_test).round(3),
        'mse_cloudcover': mean_squared_error(y_cloudcover_test, cloudcover_pred_test).round(3),
        'mse_sealevelpressure': mean_squared_error(y_sealevelpressure_test, sealevelpressure_pred_test).round(3),
        'mse_windspeed': mean_squared_error(y_windspeed_test, windspeed_pred_test).round(3),
        'mse_precip': mean_squared_error(y_precip_test, precip_pred_test).round(3),
    })

    mape_results.append({
        'name': name,
        'mape_temp': mean_absolute_percentage_error(y_temp_test, temp_pred_test).round(3),
        'mape_humidity': mean_absolute_percentage_error(y_humidity_test, humidity_pred_test).round(3),
        'mape_cloudcover': mean_absolute_percentage_error(y_cloudcover_test, cloudcover_pred_test).round(3),
        'mape_sealevelpressure': mean_absolute_percentage_error(y_sealevelpressure_test, sealevelpressure_pred_test).round(3),
        'mape_windspeed': mean_absolute_percentage_error(y_windspeed_test, windspeed_pred_test).round(3),
        'mape_precip': mean_absolute_percentage_error(y_precip_test, precip_pred_test).round(3),
    })

    # Prediksi data masa depan
    date_range = pd.date_range(start='2024-01-01', end='2030-12-30')
    future_dates = pd.DataFrame({
        'tanggal': date_range.day,
        'bulan': date_range.month,
        'tahun': date_range.year,
        'name': name
    })

    future_X = scaler_X.transform(future_dates[['tanggal', 'bulan', 'tahun']].values)
    future_X = future_X.reshape((future_X.shape[0], time_steps, 3))

    # Prediksi untuk setiap variabel pada masa depan dan denormalisasi
    future_temp = scaler_temp.inverse_transform(lstm_temp.predict(future_X)).flatten()
    future_humidity = scaler_humidity.inverse_transform(lstm_humidity.predict(future_X)).flatten()
    future_cloudcover = scaler_cloudcover.inverse_transform(lstm_cloudcover.predict(future_X)).flatten()
    future_sealevelpressure = scaler_sealevelpressure.inverse_transform(lstm_sealevelpressure.predict(future_X)).flatten()
    future_windspeed = scaler_windspeed.inverse_transform(lstm_windspeed.predict(future_X)).flatten()
    future_precip = scaler_precip.inverse_transform(lstm_precip.predict(future_X)).flatten()

    results.append({
        'name': name,
        'future_dates': date_range,
        'future_temp': future_temp,
        'future_humidity': future_humidity,
        'future_cloudcover': future_cloudcover,
        'future_sealevelpressure': future_sealevelpressure,
        'future_windspeed': future_windspeed,
        'future_precip': future_precip
    })
```

```python
mae_df = pd.DataFrame(mae_results)
mse_df = pd.DataFrame(mse_results)
mape_df = pd.DataFrame(mape_results)
mae_df.to_csv('lstm/mae.csv', index=False)
mse_df.to_csv('lstm/mse.csv', index=False)
mape_df.to_csv('lstm/mape.csv', index=False)
```

```python
predictions = []
for result in results:
    for i in range(len(result['future_dates'])):
        predictions.append({
            'name': result['name'],
            'date': result['future_dates'][i].date(),
            'temp': round(result['future_temp'][i], 3),
            'humidity': round(result['future_humidity'][i], 3),
            'cloudcover': round(result['future_cloudcover'][i], 3),
            'sealevelpressure': round(result['future_sealevelpressure'][i], 3),
            'windspeed': round(result['future_windspeed'][i], 3),
            'precip': round(result['future_precip'][i], 3)
        })

predictions_df = pd.DataFrame(predictions)
predictions_df.to_csv('lstm/prediction.csv', index=False)
```
